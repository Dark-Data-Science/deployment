+ def **PlayTimeGenre( *`genero` : str* )**:
    Debe devolver `año` con mas horas jugadas para dicho género.
  
Ejemplo de retorno: {"Año de lanzamiento con más horas jugadas para Género X" : 2013}

```python
def PlayTimeGenre(genero):
    # Filtrar el dataframe por el género específico
    df_genre = horas_usuario[horas_usuario['genres'] == genero]

    # Si no hay datos para el género, devolver un mensaje indicando esto
    if df_genre.empty:
        return f"No hay datos para el género '{genero}'."

    # Encontrar el año con más horas jugadas para el género específico
    año_con_mas_horas = df_genre.groupby('Año_estreno')['playtime_forever'].sum().idxmax()

    return año_con_mas_horas
```

```python
genero_deseado = "Adventure"  # Reemplazar con el género que estás buscando
año_con_mas_horas = PlayTimeGenre(genero_deseado)
print(f"Año de lanzamiento con más horas jugadas para Género: '{genero_deseado}' es: {año_con_mas_horas}")
```

+ def **UserForGenre( *`genero` : str* )**:
    Debe devolver el usuario que acumula más horas jugadas para el género dado y una lista de la acumulación de horas jugadas por año.

Ejemplo de retorno: {"Usuario con más horas jugadas para Género X" : us213ndjss09sdf,
			     "Horas jugadas":[{Año: 2013, Horas: 203}, {Año: 2012, Horas: 100}, {Año: 2011, Horas: 23}]}

```python
def UserForGenre(genero):
    # Filtrar el dataframe por el género específico
    df_genre = horas_usuario[horas_usuario['genres'] == genero]

    # Si no hay datos para el género, devolver un mensaje indicando esto
    if df_genre.empty:
        return {"mensaje": f"No hay datos para el género '{genero}'."}

    # Encontrar el usuario con más horas jugadas para el género específico
    usuario_con_mas_horas = df_genre.groupby('user_id')['playtime_forever'].sum().idxmax()

    # Crear una lista de la acumulación de horas jugadas por año
    acumulacion_por_anio = df_genre.groupby('Año_estreno')['playtime_forever'].sum().reset_index()
    acumulacion_por_anio = acumulacion_por_anio.rename(columns={"playtime_forever": "Horas"})
    lista_acumulacion = acumulacion_por_anio.to_dict(orient='records')

    # Crear la salida en forma de diccionario
    salida = {
        "Usuario con más horas jugadas para Género": usuario_con_mas_horas,
        "Horas jugadas": lista_acumulacion
    }

    return salida
```

```python
genero_deseado = 'Indie'  # Reemplazar con el género que estás buscando
resultado = UserForGenre(genero_deseado)

# Verificar si hay un mensaje indicando que no hay datos para el género
if 'mensaje' in resultado:
    print(resultado['mensaje'])
else:
    # Imprimir el usuario con más horas jugadas
    print(f"Usuario con más horas jugadas para Género '{genero_deseado}': Usuario: {resultado['Usuario con más horas jugadas para Género']}")
    
    # Imprimir la acumulación de horas jugadas por año
    print("Acumulación de horas jugadas por año:")
    for item in resultado['Horas jugadas']:
        print(f"Año: {item['Año_estreno']}, Horas: {item['Horas']}")

```

+ def **UsersRecommend( *`año` : int* )**:
   Devuelve el top 3 de juegos MÁS recomendados por usuarios para el año dado. (reviews.recommend = True y comentarios positivos/neutrales)
  
Ejemplo de retorno: [{"Puesto 1" : X}, {"Puesto 2" : Y},{"Puesto 3" : Z}]

```python
def UsersRecommend(recomendado_usuario, año):
    # Convertir el año a string
    año = str(año)

    # Filtrar por el año dado
    df_year = recomendado_usuario[recomendado_usuario['Año_estreno'] == año]

    # Verificar si hay datos para el año dado
    if df_year.empty:
        return f"No hay datos para el año {año}."

    # Filtrar por recomendaciones positivas y comentarios positivos/neutrales
    df_recommendations = df_year[(df_year['recommend'] == True) & (df_year['sentiment_analisis'] >= 1)]

    # Verificar si hay juegos recomendados para el año dado
    if df_recommendations.empty:
        return f"No hay juegos recomendados para el año {año}."

    # Agrupar por el título del juego y sumar el 'sentiment_analisis'
    grouped_games = df_recommendations.groupby('title')['sentiment_analisis'].sum()

    # Obtener el top 3 de juegos más recomendados
    top_3_games = grouped_games.nlargest(3)

    # Crear la lista de salida en el formato deseado
    output_list = []
    for i in range(min(3, len(top_3_games))):
        output_list.append({"Puesto {}".format(i+1): top_3_games.index[i]})

    return output_list
```

+ def **UsersWorstDeveloper( *`año` : int* )**:
   Devuelve el top 3 de desarrolladoras con juegos MENOS recomendados por usuarios para el año dado. (reviews.recommend = False y comentarios negativos)
  
Ejemplo de retorno: [{"Puesto 1" : X}, {"Puesto 2" : Y},{"Puesto 3" : Z}]

```python
def UsersWorstDeveloper(dataframe, año):
    # Convertir el año a string
    año = str(año)

    # Filtrar por el año dado
    df_year = dataframe[dataframe['Año_estreno'] == año]

    # Verificar si hay datos para el año dado
    if df_year.empty:
        return f"No hay datos para el año {año}."

    # Filtrar por recomendaciones negativas y comentarios negativos/neutrales
    df_recommendations = df_year[(df_year['recommend'] == False) & (df_year['sentiment_analisis'] <= 1)]

    # Verificar si hay juegos no recomendados para el año dado
    if df_recommendations.empty:
        return f"No hay juegos no recomendados para el año {año}."

    # Agrupar por el desarrollador del juego y sumar el 'sentiment_analisis'
    grouped_developers = df_recommendations.groupby('developer')['sentiment_analisis'].sum()

    # Obtener el top 3 de desarrolladores con juegos menos recomendados
    worst_3_developers = grouped_developers.nsmallest(3)

    # Crear la lista de salida en el formato deseado
    output_list = []
    for i in range(len(worst_3_developers)):
        output_list.append({"Puesto {}".format(i+1): worst_3_developers.index[i]})

    # Imprimir un mensaje de aviso si hay menos de 3 juegos en la salida
    if len(worst_3_developers) < 3:
        print("Advertencia: Solo se encontraron {} juegos menos recomendados para el año {}.".format(len(worst_3_developers), año))

    return output_list
```

```python
año_deseado = 2001  # Reemplazar con el año que estás buscando
resultado = UsersWorstDeveloper(recomendado_usuario, año_deseado)
print(resultado)
```

+ def **sentiment_analysis( *`empresa desarrolladora` : str* )**:
    Según la empresa desarrolladora, se devuelve un diccionario con el nombre de la desarrolladora como llave y una lista con la cantidad total 
    de registros de reseñas de usuarios que se encuentren categorizados con un análisis de sentimiento como valor. 

Ejemplo de retorno: {'Valve' : [Negative = 182, Neutral = 120, Positive = 278]}

```python
def sentiment_analysis(desarrolladora):
    # Mapeo de los valores de 'sentiment_analisis' a las cadenas de texto correspondientes
    sentiment_mapping = {0: 'Negative', 1: 'Neutral', 2: 'Positive'}

    # Filtrar el dataframe por la desarrolladora especificada
    df_developer = analisis_sentimiento[analisis_sentimiento['developer'] == desarrolladora]

    # Si no hay datos para la desarrolladora, devolver un mensaje indicando esto
    if df_developer.empty:
        return {desarrolladora: "No hay datos para la desarrolladora."}

    # Reemplazar los valores en 'sentiment_analisis' con las cadenas de texto correspondientes
    df_developer['sentiment_analisis'] = df_developer['sentiment_analisis'].map(sentiment_mapping)

    # Contar el número de registros de reseñas de usuarios que se encuentran categorizados con un análisis de sentimiento
    sentiment_counts = df_developer['sentiment_analisis'].value_counts().to_dict()

    # Crear la lista de salida en el formato deseado
    output_list = [f"{key} = {value}" for key, value in sentiment_counts.items()]

    return {desarrolladora: output_list}

```

```python
# Ingresar el nombre de la empresa desarrolladora
desarrolladora = "SEGA"  # Reemplazar con el nombre de la empresa desarrolladora que estás buscando

# Llamar a la función con el nombre de la empresa desarrolladora
resultado = sentiment_analysis(desarrolladora)

# Imprimir el resultado
print(resultado)

```
